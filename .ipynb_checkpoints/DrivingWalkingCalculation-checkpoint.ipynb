```python
import numpy as np
import pandas as pd
import geopandas as gpd
import networkx as nx
import osmnx as ox
```

```python
df = pd.read_csv("realtor_data.csv")
df
```

```python
place = 'Tompkins County, New York, USA'
G = ox.graph_from_place(place)
fig, ax = ox.plot_graph(G)
```

```python
G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)
```

```python
orig = ox.nearest_nodes(G, -76.484735,42.450335)
dest_cornell = ox.nearest_nodes(G, -76.48274160938676, 42.44757971297041)
route = ox.shortest_path(G, orig, dest_cornell, weight='travel_time')
fig, ax = ox.plot_graph_route(G, route, node_size=0)
```

```python
int(sum(ox.routing.route_to_gdf(G, route, weight="travel_time")["travel_time"]))
```
